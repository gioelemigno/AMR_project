```python
FOLDER_IN_INPUTS = 'inputs'
FOLDER_IN_WORKSPACE = 'traj'
FOLDER_IN_V_2D = 'v_2d'
FOLDER_IN_V_3D = 'V_comp'

FOLDER_OUT = 'OUT_FRAMES'
```

```python
def crop_workspace_or_C_space(file_in):
    im = Image.open(file_in)

    width, height = im.size

    left = width/4
    top = 0
    right = 3*width/4
    bottom = height

    im1 = im.crop((left, top, right, bottom))
    #im1.show()
    #im1.save("out1.png")
    return im1

def crop_control(file_in):
    im = Image.open(file_in)

    width, height = im.size

    left = 0
    top = 0
    right = width / 2
    bottom = height

    im1 = im.crop((left, top, right, bottom))
    #im1.show()
    #im1.save("out1.png")
    return im1

def crop_disturbance(file_in):
    im = Image.open(file_in)

    width, height = im.size

    left = width / 2 
    top = 0
    right = width 
    bottom = height - 2/3 * height;

    im1 = im.crop((left, top, right, bottom))
    #im1.show()
    #im1.save("out1.png")
    return im1

def crop_V3D(file_in):
    im = Image.open(file_in)
    return im
    '''
    width, height = im.size

    left = width / 2 
    top = 0
    right = width 
    bottom = height - height/3

    im1 = im.crop((left, top, right, bottom))
    #im1.show()
    #im1.save("out1.png")
    return im1
    '''
```

```python
def resize_workspace(img_in, width_bg, height_bg):
    #img_in = workspace

    width, height = img_in.size
    ratio = width/height 

    wo = width_bg / 2.7
    ho = wo / ratio

    wo = int(wo)
    ho = int(ho)

    im = img_in.resize((wo, ho), resample=Image.LANCZOS) # PIL.Image.LANCZOS (a high-quality downsampling filter)
    #display(im)
    return im
    
def resize_c_space(img_in, width_bg, height_bg):
    #img_in = c_space

    width, height = img_in.size
    ratio = width/height 

    wo = width_bg / 2.7
    ho = wo / ratio

    wo = int(wo)
    ho = int(ho)

    im = img_in.resize((wo, ho), resample=Image.LANCZOS) # PIL.Image.LANCZOS (a high-quality downsampling filter)
    #display(im)
    return im

def resize_V3D(img_in, width_bg, height_bg):
    #img_in = c_space

    width, height = img_in.size
    ratio = width/height 

    wo = width_bg / 4
    ho = wo / ratio

    wo = int(wo)
    ho = int(ho)

    im = img_in.resize((wo, ho), resample=Image.LANCZOS) # PIL.Image.LANCZOS (a high-quality downsampling filter)
    #display(im)
    return im

def resize_u(img_in, width_bg, height_bg):
    #img_in = c_space

    width, height = img_in.size
    ratio = width/height 

    wo = width_bg / 4
    ho = wo / ratio

    wo = int(wo)
    ho = int(ho)

    im = img_in.resize((wo, ho), resample=Image.LANCZOS) # PIL.Image.LANCZOS (a high-quality downsampling filter)
    #display(im)
    return im

def resize_d(img_in, width_bg, height_bg):
    #img_in = c_space

    width, height = img_in.size
    ratio = width/height 

    wo = width_bg / 4
    ho = wo / ratio

    wo = int(wo)
    ho = int(ho)

    im = img_in.resize((wo, ho), resample=Image.LANCZOS) # PIL.Image.LANCZOS (a high-quality downsampling filter)
    #display(im)
    return im
```

```python
def place_c_space(img_in, background, BUTTON_LABEL_SIZE):
    #img_in = im
    #background = background

    width_bg, height_bg = background.size
    width_img, height_img = img_in.size 

    res = background.copy()

    left = width_bg - width_img 
    top = height_bg - height_img - BUTTON_LABEL_SIZE
    res.paste(img_in, (int(left), int(top)))
    #display(res)
    return res

def place_workspace(img_in, background, BUTTON_LABEL_SIZE):
    #img_in = im
    #background = background

    width_bg, height_bg = background.size
    width_img, height_img = img_in.size 

    res = background.copy()

    left = width_bg - 2 * width_img
    top = height_bg - height_img - BUTTON_LABEL_SIZE
    res.paste(img_in, (int(left), int(top)))
    #display(res)
    return res

def place_V3D(img_in, background, BUTTON_LABEL_SIZE):
    #img_in = im
    #background = background

    width_bg, height_bg = background.size
    width_img, height_img = img_in.size 

    res = background.copy()

    left = width_bg - width_img - width_img/5
    top = 0
    res.paste(img_in, (int(left), int(top)))
    #display(res)
    return res

def place_u(img_in, background, BUTTON_LABEL_SIZE):
    #img_in = im
    #background = background

    width_bg, height_bg = background.size
    width_img, height_img = img_in.size 

    res = background.copy()

    left = 0#width_bg - 2 * width_img
    top = height_bg - 1.8 * height_img - BUTTON_LABEL_SIZE
    res.paste(img_in, (int(left), int(top)))
    #display(res)
    return res

def place_d(img_in, background, BUTTON_LABEL_SIZE):
    #img_in = im
    #background = background

    width_bg, height_bg = background.size
    width_img, height_img = img_in.size 

    res = background.copy()

    left = width_img/15
    top = height_bg - height_img - height_img - BUTTON_LABEL_SIZE
    res.paste(img_in, (int(left), int(top)))
    #display(res)
    return res
```

```python
def add_lines(img_in):
    width_bg, height_bg = img_in.size

    res = ImageDraw.Draw(img_in)

    x0 = width_bg/3.9
    y0 = height_bg / 10
    #
    x1 = x0
    y1 = height_bg
    res.line([(int(x0), int(y0)), (int(x1), int(y1))], fill=(200,200,200), width=4)

    x0 = width_bg * 1.9/3
    y0 = height_bg/10
    #
    x1 = x0
    y1 = height_bg
    res.line([(int(x0), int(y0)), (int(x1), int(y1))], fill=(200, 200, 200), width=4)

    return res._image


def add_text(img_in, plot_time):
    width_bg, height_bg = img_in.size

    res = ImageDraw.Draw(img_in)

    x = width_bg * 0.06
    y = height_bg * 0.90
    text = "PLAYER MOVES"
    font = ImageFont.truetype('./Open_Sans/static/OpenSans/OpenSans-Medium.ttf', 60)
    res.text((int(x), int(y)), text=text, font=font, fill=(0,0,0))

    x = width_bg * 0.38
    text = "WORKSPACE"
    res.text((int(x), int(y)), text=text, font=font, fill=(0,0,0))

    x = width_bg * 0.75
    text = "C-SPACE (3D|2D)"
    res.text((int(x), int(y)), text=text, font=font, fill=(0,0,0))

    x = width_bg * 0.43
    y = height_bg * 0.05

    font = ImageFont.truetype('./Open_Sans/static/OpenSans/OpenSans-Medium.ttf', 80)
    text = plot_time
    res.text((int(x), int(y)), text=text, font=font, fill=(0,0,0))

    return res._image
```

```python
def create_frame(file_workspace, file_v2d, file_v3d, file_inputs):
    workspace = crop_workspace_or_C_space(file_workspace)
    c_space = crop_workspace_or_C_space(file_v2d)
    v_3d = crop_V3D(file_v3d)
    u = crop_control(file_inputs)
    d = crop_disturbance(file_inputs)

    plot_time = file_workspace[file_workspace.find('_t=')+1: file_workspace.find('.png')] 
    
    im = Image.open(file_workspace)
    width_bg, height_bg = im.size
    BUTTON_LABEL_SIZE = height_bg/6
    height_bg += BUTTON_LABEL_SIZE

    width_bg = width_bg + 1 if width_bg % 2 != 0 else width_bg
    height_bg = height_bg + 1 if height_bg % 2 != 0 else height_bg

    background = Image.new("RGB", (int(width_bg), int(height_bg)), (255, 255, 255))

    workspace = resize_workspace(workspace, width_bg, height_bg)
    c_space = resize_c_space(c_space, width_bg, height_bg)
    v_3d = resize_V3D(v_3d, width_bg, height_bg)
    u = resize_u(u, width_bg, height_bg)
    d = resize_d(d, width_bg, height_bg)

    res = place_workspace(workspace, background, BUTTON_LABEL_SIZE)
    res = place_c_space(c_space, res, BUTTON_LABEL_SIZE)
    res = place_u(u, res, BUTTON_LABEL_SIZE)
    res = place_d(d, res, BUTTON_LABEL_SIZE)
    res = place_V3D(v_3d, res, BUTTON_LABEL_SIZE)

    res = add_lines(res)
    res = add_text(res, plot_time)

    return res
```

```python
def create_frames_folder(root_folder):
    folder_in_inputs = os.path.join(root_folder, FOLDER_IN_INPUTS)
    list_file_inputs = sorted(os.listdir(folder_in_inputs))
    list_file_inputs = list(filter(lambda x: '.png' in x, list_file_inputs))
    list_file_inputs = list_file_inputs + [list_file_inputs[-1]]
    list_file_inputs = [list_file_inputs[0]]*9 + list_file_inputs + [list_file_inputs[-1]]*9

    folder_in_v_2d = os.path.join(root_folder, FOLDER_IN_V_2D)
    list_file_v2d = sorted(os.listdir(folder_in_v_2d))
    list_file_v2d = list(filter(lambda x: '.png' in x, list_file_v2d))
    list_file_v2d = [list_file_v2d[0]]*9 + list_file_v2d + [list_file_v2d[-1]]*9

    folder_in_v3d = os.path.join(root_folder, FOLDER_IN_V_3D)
    list_file_v3d = sorted(os.listdir(folder_in_v3d), reverse=True)
    list_file_v3d = list(filter(lambda x: '.png' in x, list_file_v3d))
    list_file_v3d = list_file_v3d[1:] + [list_file_v3d[0]]
    list_file_v3d = [list_file_v3d[0]]*9 + list_file_v3d + [list_file_v3d[-1]]*9

    folder_in_workspace = os.path.join(root_folder, FOLDER_IN_WORKSPACE)
    list_file_workspace = sorted(os.listdir(folder_in_workspace))
    list_file_workspace = list(filter(lambda x: '.png' in x, list_file_workspace))
    list_file_workspace = [list_file_workspace[0]]*9 + list_file_workspace + [list_file_workspace[-1]]*9

    folder_out = os.path.join(root_folder, FOLDER_OUT)
    if not os.path.isdir(folder_out):
        os.makedirs(folder_out)

    for i, (file_inputs, file_v2d, file_v3d, file_workspace) in enumerate(zip(list_file_inputs, list_file_v2d, list_file_v3d, list_file_workspace)):
        file_workspace = os.path.join(folder_in_workspace, file_workspace)
        file_v2d = os.path.join(folder_in_v_2d, file_v2d)
        file_v3d = os.path.join(folder_in_v3d, file_v3d)
        file_inputs = os.path.join(folder_in_inputs, file_inputs)
        
        res = create_frame(file_workspace, file_v2d, file_v3d, file_inputs)
        filename = 'frame_' + str(i).zfill(4) + '.png'

        res.save(os.path.join(folder_out, filename))
```

```python
def get_list_dirs_experiments(current_folder, list_out, required_folders):
    items = os.listdir(current_folder)
    subfolders = list(filter(lambda x: os.path.isdir(os.path.join(current_folder, x)), items))

    there_are_all_req = sum([1 if req_f in subfolders else 0 for req_f in required_folders]) == len(required_folders)
    if there_are_all_req:
        list_out.append(current_folder)
    
    for folder in subfolders:
        new_current_folder = os.path.join(current_folder, folder)
        list_out = get_list_dirs_experiments(new_current_folder, list_out, required_folders)

    return list_out
```

```python
import subprocess

# ffmpeg -framerate 3 -i ./OUT_FRAMES/frame_%04d.png -c:v libx264 -profile:v high -crf 20 -pix_fmt yuv420p output.mp4

folders = get_list_dirs_experiments('.', [], 
                    required_folders=[FOLDER_IN_INPUTS, FOLDER_IN_V_2D, FOLDER_IN_V_3D, FOLDER_IN_WORKSPACE])
for folder in folders:
    print(folder)
    create_frames_folder(folder)

    cmd = 'ffmpeg '
    cmd += '-y ' # overwrite
    cmd += '-framerate 3 -i '
    cmd += os.path.join(folder, FOLDER_OUT, 'frame_%04d.png')
    cmd += ' -c:v libx264 -profile:v high -crf 20 -pix_fmt yuv420p '
    cmd += os.path.join(folder, FOLDER_OUT, 'output.mp4') 
    results = subprocess.run(cmd, shell=True, universal_newlines=True, check=True)
    #break
```
